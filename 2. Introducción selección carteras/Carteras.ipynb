## FASES DE LA SELECCIÓN DE CARTERAS
- Determinar **rentabilidad esperada, riesgo y correlación** de los activos.
- Determinar **rentabilidad esperada y riesgo** de las carteras.
- Determinar **cartera óptima** dado un objetivo de rentabilidad o de riesgo esperado.
- Seguimiento de la evolución del rendimiento de la cartera

### **Ejemplo 2.1.**

Indica el orden de elección entre los activos

```python
import pandas as pd
import matplotlib.pyplot as plt

act=pd.DataFrame({"Activos": ["A","B","C","D","E","F"],
"Rentabilidad Esperada": [6,11,9,7,9,10],
"Riesgo": [2,7,5,2,4,6]})

plt.scatter(act["Riesgo"], act["Rentabilidad Esperada"])
plt.ylabel("Rentabilidad esperada")
plt.xlabel("Riesgo")
# Añadir etiquetas correctas
for activo, rentabilidad, riesgo in zip(act["Activos"], act["Riesgo"], act["Rentabilidad Esperada"]):
    plt.annotate(activo, #etiqueta
                (rentabilidad, riesgo), #posición
                xytext=(5,5), #texto desplazado 5 píxeles a la derecha y arriba
                textcoords="offset points") #el texto desplazado es desplazado con respecto al punto, no en coordenadas del gráfico.


plt.show()
```

- E es preferido a C
- D es preferido a A
- Para elegir entre D, E, F, B cada invesor lo hará en base a sus preferencias rentabilidad-riesgo

Si no se conoce el grado de aversión al riesgo del inversor, se eligirá aquella cartera que minimice el **coficiente de variación**:

$$cv=\frac{\sigma_i}{E_i}$$

Este cociente cuantifica el nivel de riesgo por cada unidad de rentabilidad.

```python
# Crea una lista de los valores que quieres eliminar
act["CV"]=act["Riesgo"]/act["Rentabilidad Esperada"]

plt.scatter(act["Activos"], act["CV"])
```

Por lo tanto, el activo a elegir sería D pues es el que miniza el riesgo por cada unidad de rentabilidad.
